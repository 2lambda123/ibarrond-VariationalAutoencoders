```python
# uses gradient clipping
# 10 mc samples
# with variational ae kl
n_inputs = 784
n_neurons_encoder = [256, 128]
n_latent = 2
n_neurons_decoder = [128, 256]
vi = BayesianVAEArtificial('BAE_256_128_2',
                         n_inputs,
                         n_neurons_encoder,
                         n_latent,
                         n_neurons_decoder,
                         constant_prior=False)
vi.learn(learning_rate=0.01, epochs=50, batch_size=64, mc_samples=20)
print('Test Log-Likelihood: %.6f' % vi.benchmark())
```
