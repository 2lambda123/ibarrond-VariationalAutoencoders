```python
import tensorflow as tf
```

```python
sess = tf.Session()

X = tf.round(tf.random_normal(shape=[1, 3, 2]) * 10)

reduced = tf.reduce_sum(X, reduction_indices=[-1])
meaned = tf.reduce_mean(reduced, 0)

X, reduced, meaned = sess.run([X, reduced, meaned])

print('X')
print(X)
print('Reduced')
print(reduced)
print('Meaned')
print(meaned)
```

```python


sess = tf.Session()

mc_samples = 5
batch_size = 3
d_in = 2
d_out = 3

X = tf.round(tf.random_normal(shape=[batch_size, d_in]) * 10)
inputs = tf.multiply(tf.ones([mc_samples, batch_size, d_in]), X)

weights = tf.round(tf.random_normal(shape=[mc_samples, d_in+1, d_out]) * 10)
bias = tf.expand_dims(weights[:,0,:], 1)

outputs = tf.matmul(inputs, weights[:,1:,:])
outputs_bias = outputs + bias

mul = tf.reshape(tf.cast(tf.range(9), tf.float32), [3,3])

mul_shift = tf.add(tf.multiply(outputs_bias, mul), 1)

reduced = tf.reduce_sum(outputs_bias, reduction_indices=[-1])
reduced_mean = tf.reduce_mean(reduced, 0)

#ones = tf.ones([2,3,2])
#zeros = tf.zeros([3,3,2])

#base = tf.concat([ones, zeros], 0)

#rnd = tf.sigmoid(base)

inputs, weights, bias, outputs, outputs_bias, mul_shift, reduced, reduced_mean = sess.run([inputs, weights, bias, outputs, outputs_bias, mul_shift, reduced, reduced_mean])

print('Inputs')
print(inputs)
print('Weights')
print(weights)
print('Bias')
print(bias)
print('Outputs')
print(outputs)
print('Outputs incl. Bias')
print(outputs_bias)
print('Mul Shift')
print(mul_shift)
print('Reduced')
print(reduced)
print('Reduced Mean')
print(reduced_mean)
```
