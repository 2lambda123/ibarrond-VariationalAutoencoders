# Adversarial Autoencoder

<div class="alert alert-info">
<p>In this notebook we experiment with an adversarial autoencoder. The latent distribution is constrained to be gaussian using an adversarial training procedure.</p>
</div>

```python
from models.GAN import AAE
```

# 32

```python
my_aae = AAE('AAE_32', n_inputs=784,
             n_neurons_encoder=[],
             n_latent=32,
             n_neurons_decoder=[])
my_aae.learn(learning_rate=0.001, n_epochs=50)
my_aae.serialize('DUMPS/' + my_aae.name + '/model.ckpt')
print('Test Log-Likelihood: %.6f' % my_aae.benchmark())
```

```python
my_aae.plot_enc_dec(20, save=True)
```

```python
my_aae.plot_noisy_recon(20, save=True)
```

```python
my_aae.benchmark(noisy=True)
```
